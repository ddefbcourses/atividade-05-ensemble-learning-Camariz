# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `MNIST handwritten digits` (arquivos de treino e teste)
Realize a separação do conjunto de treino como treino e validação
Utilize `train_test_split` com controle de aleatoriedade (seed)
Retorne: `X_train`, `X_val`, `y_train`, `y_val`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

```python
# TODO: implemente load_data
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("mnist_train.csv")
```

```python
def load_data(seed):
    X = df.drop("label", axis=1)
    y = df["label"]

    X_train, X_val, y_train, y_val = train_test_split(
        X,
        y,
        test_size=0.2,
        random_state=seed,
        stratify=y
    )

    return X_train, X_val, y_train, y_val
```

```python
X_train, X_val, y_train, y_val = load_data(42)

print(X_train.shape)
print(X_val.shape)
```

```python
X_train = X_train / 255.0
X_val = X_val / 255.0
```

```python
X_train.head()
```

```python
X_train.min().min(), X_train.max().max()
```

Sim, é necessário normalizar os dados. O dataset MNIST possui valores de pixels entre 0 e 255. Essa diferença de escala pode prejudicar o desempenho de algoritmos de machine learning. A normalização coloca os valores na faixa de 0 a 1, o que melhora a estabilidade numérica, acelera o treinamento e melhora o desempenho dos modelos.

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

```python
# TODO: implemente load_data
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def train_random_forest(X_train, y_train, seed):

    model = RandomForestClassifier(
        n_estimators=100,
        random_state=seed,
        n_jobs=-1
    )

    model.fit(X_train, y_train)

    return model
```

```python
model = train_random_forest(X_train, y_train, 42)

print(model)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_val)
accuracy_score(y_val, y_pred)
```

O modelo Random Forest obteve uma acurácia de 96,65% na validação. Esse resultado indica um bom desempenho do modelo na classificação dos dígitos manuscritos do dataset MNIST. A performance pode ser atribuída à capacidade do Random Forest de combinar múltiplas árvores de decisão, reduzindo overfitting e aumentando a generalização.

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

```python
# TODO: implemente load_data
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
def run_pipeline(model_type="rf", seed=42):

    X_train, X_val, y_train, y_val = load_data(seed)


    if model_type == "rf":
        model = RandomForestClassifier(
            n_estimators=100,
            random_state=seed,
            n_jobs=-1
        )
    else:
        raise ValueError("Modelo não suportado")

    
    model.fit(X_train, y_train)

    
    y_pred = model.predict(X_val)
    acc = accuracy_score(y_val, y_pred)

    return acc
```

```python
run_pipeline("rf", 42)
```

O valor de acurácia permaneceu consistente devido à reprodutibilidade do experimento. O uso do parâmetro random_state garante que a divisão dos dados e o treinamento do modelo sejam determinísticos, produzindo o mesmo resultado para a mesma configuração.

```python
run_pipeline("rf", seed=7)
```

A pequena variação na acurácia ocorre devido às diferentes divisões do conjunto de dados durante o treino e validação. Mesmo com o uso de random_state, pequenas mudanças na distribuição dos dados podem impactar o desempenho do modelo, especialmente em algoritmos baseados em ensembles como o Random Forest.

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

O overfitting em árvores de decisão tende a começar quando a profundidade do modelo se torna muito alta, geralmente a partir de valores elevados (por exemplo, acima de 10–20 níveis), dependendo da complexidade dos dados. Quando max_depth=None, a árvore não possui restrição de profundidade e cresce até que todas as folhas sejam puras ou contenham o número mínimo de amostras. Isso faz com que o modelo memorize completamente o conjunto de treino, resultando em 100% de acurácia no treino. No entanto, isso prejudica a capacidade de generalização, caracterizando overfitting.

A acurácia é uma métrica importante e bastante intuitiva, pois indica a proporção geral de acertos do modelo. No contexto do MNIST, ela já fornece uma boa visão inicial do desempenho, especialmente porque o problema é relativamente balanceado e bem definido. No entanto, a acurácia sozinha não é suficiente para uma avaliação completa. Ela não captura possíveis erros por classe nem revela desequilíbrios no comportamento do modelo. Por isso, métricas como precisão, recall e F1-score são necessárias para uma análise mais robusta e confiável.

Para garantir que os resultados não ocorreram por acaso, é fundamental utilizar estratégias de reprodutibilidade e validação. O uso do parâmetro random_state assegura que as divisões de dados e o treinamento sejam consistentes entre execuções, permitindo replicação dos experimentos. Além disso, a utilização de um conjunto de validação separado e a análise de diferentes seeds ajudam a verificar a estabilidade do modelo. Caso os resultados permaneçam consistentes em diferentes execuções, isso aumenta a confiança de que o desempenho observado não foi fruto do acaso.

Um possível problema metodológico é a ausência de validação cruzada (cross-validation), que poderia fornecer uma estimativa mais robusta do desempenho do modelo ao invés de depender de uma única divisão treino/validação. Outro problema é a possibilidade de overfitting, especialmente em modelos baseados em árvores com alta profundidade, como observado quando max_depth=None. Isso pode levar o modelo a memorizar o conjunto de treino, reduzindo sua capacidade de generalização.

O pipeline implementado é confiável do ponto de vista técnico, pois segue boas práticas como separação entre treino e validação, uso de reprodutibilidade com random_state, normalização dos dados e avaliação com múltiplas métricas.

No entanto, sua confiabilidade poderia ser ainda maior com a inclusão de validação cruzada, testes em diferentes particionamentos dos dados e comparação com outros modelos. Dessa forma, embora o pipeline seja sólido e consistente para fins educacionais e experimentais, ainda há espaço para torná-lo mais robusto em um cenário de produção.
